```python
#Problem1
#import libraries
from collections import Counter 
from nltk.tokenize import sent_tokenize, word_tokenize
import matplotlib.pylab as plt
import matplotlib.pyplot as p

def zipfs_plot(corpus_path):
    """
    This is a function, which tokenizes corpora, creates frequency lists and plots the frequency of words
    
    Parameters: 
    corpus_path(str): it takes the absolute path for the corpus

    Return:
    None
    """
    #read the corpus
    data = open(corpus_path,'r').read()
    data = data.replace("\n","")
    tokens_sentences =sent_tokenize(data) 
    tokens = []

    #tokenize the corpus in sentences and words    
    for each in tokens_sentences:
        tokens_w = word_tokenize(each)
        tokens += tokens_w
    
    #create a list for word frequence
    frequency_of_words = Counter(tokens)
    new_list = {key: value for key, value in sorted(frequency_of_words.items(), key=lambda item: item[1], reverse=True)}
    frequencies = list(new_list.values())
    
    #plot the frequency of words in linear and logaritmic scales
    p.plot(frequencies)
    p.xlabel("Rank of Words")
    p.ylabel("Absolute Frequency")
    p.show()
    p.loglog(frequencies)
    p.xlabel("Rank of Words(log)")
    p.ylabel("Frequency(log)")
    p.show()
```

```python
zipfs_plot("junglebook.txt")
```

```python
zipfs_plot("kingjamesbible.txt")
```

```python
zipfs_plot("SETIMES.bg-tr.bg")
```

```python
zipfs_plot("SETIMES.bg-tr.tr")
```

## Findings ##
Plotting the junglebook-corpus and the kingjamesbible-corpus on a log-log graph, we can verify that both corpora follow Zipf's Law. The graphs indicate that in both corpora words with high frequency of occurrence are assigned low ranks, whereas words with low frequency of occurence are assigned high ranks. That means few words occur frequently, whereas many words appear rarely. 

The Jungle Book is a relatively small corpus, so that the most frequent words occur only about 4000 times. The rapid drop of the curves to values between zero and about one hundred for the remaining almost 7500 ranks, shows how small the number of frequently occurring words and how high the number of rarely occurring words actually is. Looking at the kingjamesbible-corpus the differences in the frequency distribution become even more obvious: The most frequent words occur around 70 000 times, while the vast majority of words with a frequency of less than 100 occurrences account for almost all of the remaining 14 000 ranks.The big distribution differences can be better observed in the logarithmic scale, whereas the inverse relation between the probability of the occurence of a word and its ranking in the frequency order is indicated by the curves in the linear scales, which resemble hyperboles.
One reason for this might be the bigger corpus size of the bible. Another, probably more important reason, might be, that the Jungle Book is a children's book with a limited vocabulary, while the King James Bible contains a broad vocabulary with many rare, sometimes outdated words.   

By plotting the Turkish-Bulgarian parallel corpus on a log-log graph we can see, that Zipf's law applies for these two corpora as well. Again, words with high frequency of occurrence are assigned low ranks, whereas words with low frequency of occurence are assigned high ranks.                                                             Although the curves look the same at first glance, a closer look reveals differences. The occurrence of the most frequent words in the Bulgarian text is much higher than the occurrence of the most frequent words in the Turkish text: While the most frequent words in the Bulgarian text appear approximately 30 000 times, the words, that are the most frequent in the Turkish text occur around 23 000 times. At the same time, the frequencies of the words in the Turkish corpus are distributed over around 300 000 ranks, whereas in Bulgarian they are only distributed over approximately 250 000 ranks.

This may be due to the different structure of the two languages: Bulgarian is an inflectional language, Turkish is an agglutinative language, which has no prepositions, no articles and no grammatical gender. This plays a role for the findings since in most of the indoeuropean languages, to which Bulgarian belongs, the words with the highest frecuency of occurence are normally function words. In Turkish, grammatical functions can only be indicated by the affixes attached to content words. Since these can be combined in million different ways, the probability that exactly the same constellations will occur again and again is more unlikely than the occurence of the same (grammar) words in Bulgarian.
